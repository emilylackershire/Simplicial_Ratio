```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle as pkl
import Hypergraph_Functions as hf
```

```python
with open('adversarial_experiment.pkl', 'rb') as file:
    df = pkl.load(file)
df
```

```python
n = {}
m = {}
datasets = [
    "contact-primary-school",
    "contact-high-school",
    "hospital-lyon",
    "email-enron",
    "email-eu",
    "diseasome",
    "disgenenet",
    "ndc-substances",
#    "congress-bills",
#    "ubuntu (edge-chopped)",
]
for dataset in datasets:
    n[dataset] = df['n'][dataset]
    m[dataset] = df['m'][dataset]
    print(dataset)
    print(n[dataset], m[dataset])
```

```python
experiments = ["adversarial growth"]
limited_datasets = [
#    "contact-primary-school",
#    "contact-high-school",
    "hospital-lyon",
#    "email-enron",
#    "email-eu",
#    "diseasome",
    "disgenenet",
#    "ndc-substances",
#    "congress-bills",
#    "ubuntu (edge-chopped)",
]
for d in limited_datasets:
    stop_value = 0.9*n[d]
    print('stop_value for {0} is {1}'.format(d, stop_value))
    for e in experiments:
        c = 0
        while True:
            if df[e][d][c] < stop_value:
                c += 1
            else:
                break
        cutoff[d][e] = c+1
print(cutoff)
```

```python
names = [
    "contact-primary-school",
    "contact-high-school",
    "hospital-lyon",
    "email-enron",
    "email-eu",
    "diseasome",
#    "disgenenet",
#    "ndc-substances",
#    "congress-bills",
#    "ubuntu (edge-chopped)",
#    "ubuntu (vertex-chopped)"
]

experiments = [
    "random growth",
    "single-source diffusion",
    "10% sprinkled diffusion",
]

for i, experiment in enumerate(experiments):
    fig, axs = plt.subplots(3, 2, figsize = (10,12))
    row = 0
    col = 0
    for name in names: 
        if i<1:
            axs[row, col].plot(list(df[experiment][name].keys())[:cutoff[name][experiment]], 
                               [x/n[name] for x in df[experiment][name].values()][:cutoff[name][experiment]], 
                               label='Real', 
                               linewidth=2, 
                               color='black'
                              )
    #        plt.plot(df[experiment]['{0} (Chung-Lu)'.format(name)].keys(),df[experiment]['{0} (Chung-Lu)'.format(name)].values(),label='Chung-Lu')
            axs[row, col].plot(list(df[experiment]['{0} (q=0)'.format(name)].keys())[:cutoff[name][experiment]], 
                               [x/n[name] for x in df[experiment]['{0} (q=0)'.format(name)].values()][:cutoff[name][experiment]],
                               label='q=0'
                              )
            axs[row, col].plot(list(df[experiment]['{0} (q=0.5)'.format(name)].keys())[:cutoff[name][experiment]], 
                               [x/n[name] for x in df[experiment]['{0} (q=0.5)'.format(name)].values()][:cutoff[name][experiment]], 
                               label='q=0.5')
            axs[row, col].plot(list(df[experiment]['{0} (q=1)'.format(name)].keys())[:cutoff[name][experiment]], 
                               [x/n[name] for x in df[experiment]['{0} (q=1)'.format(name)].values()][:cutoff[name][experiment]], 
                               label='q=1')
    #        plt.plot(df[experiment]['{0} (bottom-up)'.format(name)].keys(),df[experiment]['{0} (bottom-up)'.format(name)].values(),label='Bottom-up simplicial')
    #        plt.plot(df[experiment]['{0} (top-down)'.format(name)].keys(),df[experiment]['{0} (top-down)'.format(name)].values(),label='Top-down simplicial')            
        else:
            axs[row, col].plot(df[experiment][name].keys(), df[experiment][name].values(), label='Real', linewidth=2, color='black')
    #        plt.plot(df[experiment]['{0} (Chung-Lu)'.format(name)].keys(),df[experiment]['{0} (Chung-Lu)'.format(name)].values(),label='Chung-Lu')
            axs[row, col].plot(df[experiment]['{0} (q=0)'.format(name)].keys(),df[experiment]['{0} (q=0)'.format(name)].values(),label='q=0')
            axs[row, col].plot(df[experiment]['{0} (q=0.5)'.format(name)].keys(),df[experiment]['{0} (q=0.5)'.format(name)].values(),label='q=0.5')
            axs[row, col].plot(df[experiment]['{0} (q=1)'.format(name)].keys(),df[experiment]['{0} (q=1)'.format(name)].values(),label='q=1')
    #        plt.plot(df[experiment]['{0} (bottom-up)'.format(name)].keys(),df[experiment]['{0} (bottom-up)'.format(name)].values(),label='Bottom-up simplicial')
    #        plt.plot(df[experiment]['{0} (top-down)'.format(name)].keys(),df[experiment]['{0} (top-down)'.format(name)].values(),label='Top-down simplicial')
        if row == 0 and col == 0:
            axs[row, col].legend()
        if col == 0:
            if i<1:
                axs[row, col].set_ylabel('Normalized size of largest component')
            else:
                axs[row, col].set_ylabel('Wasserstein distnace')
        if row == 2:
            if i<1:
                axs[row, col].set_xlabel('Number of edges added')
            else:
                axs[row, col].set_xlabel('Round')
        axs[row, col].set_title('{0}'.format(name))
        if col == 1:
            col = 0
            row += 1
        else:
            col += 1
    plt.show()
    fig.savefig('pictures/first-six-{0}.png'.format(experiment))
    plt.close()
```
